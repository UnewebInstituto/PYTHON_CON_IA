```python
# Declaración de dependencia
import numpy as np
# Declaración de lista de datos
datos = [25,37,42,16,88,23,26]
# Calcular la media arítmetica
promedio = np.mean(datos)
# Calcular la desviación estándar
desviacion = np.std(datos)
# Imprimir resulatdos
print(f"El promedio de los datos es {promedio}")
print(f"La desviación estándar es {desviacion}")
```
